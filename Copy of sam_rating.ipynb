```python
#from the csv now we want to extract only columns with ratings

columns=list(range(2, 3))
ratings = pd.read_csv('G1.csv', usecols=columns, skiprows =7 , nrows=5)
print(ratings) 
```

```python
# JUST FOR VISUALIZATION

# getting a column name
name=list(ratings.columns)
#soriting it so that we get the highest value first, and we want it to be in descending order
print(ratings.sort_values(name, ascending=False))
```

```python
# creating a dictionary with ratings and the indeces
dict1 = ratings.to_dict()
#print(dict1)
#print()
lista2=list(dict1.values())
#print(lista2)
our_dictionary={}
for element in lista2:
    for key in element:
        #print(type( element[key]))
        new_element = int(element[key].replace('.', ''))
        our_dictionary[key] =new_element

print(our_dictionary)
```

```python
#getting the maximum rating index
max_key = max(our_dictionary, key=our_dictionary.get)
# print(max_key)

# funny -> 1 -> 4 row, 0 row, 3 row
#boring -> 2 -> 1 row
#horrible -> 3 -> 2 row

if max_key == 0 or max_key == 3 or max_key == 4:
    our_value_for_37nparray= 1
elif max_key == 1:
    our_value_for_37nparray= 2
elif max_key == 2:
    our_value_for_37nparray=3
```
